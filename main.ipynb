```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# load dataset
df = pd.read_csv("perfume.csv")
df.head()
```

```python
df.info()
```

```python
# drop duplicate data
duplicate = df[df.duplicated()]
print(duplicate.shape)

df = df.drop_duplicates()
```

```python
df.count()
```

```python
print(df.isnull().sum())
```

```python
# display the data(barplot)
new_df = df.head(20)
plt.bar(new_df["brand"], new_df["followers"], color="red", width=0.5)
plt.xlabel("Indonesian Perfume Brand")
plt.ylabel("Followers")
plt.title("Followers data on Indonesian perfume brands")
plt.xticks(rotation=45)
plt.show()
```

```python
new_df = df.head(10)
plt.pie(new_df["followers"], labels=new_df["brand"])
```

```python
# displaying with scatter plot
new_df=df.head(20)
plt.scatter(new_df["followers"], new_df["price_per_ml"])
plt.xlabel("Followers")
plt.ylabel("Price per ML")
plt.title("Scatter plot of Followers vs Price per ML")
plt.show()
```

```python
# showing outlier
outl = plt.figure(figsize=(12,4))
outl.add_subplot(1,2,1)
df["followers"].plot(kind="area")
outl.add_subplot(1,2,2,)
plt.boxplot(df["followers"])
plt.show()
```
